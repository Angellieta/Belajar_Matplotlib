```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
# MEMBUAT PLOTTING SEDERHANA
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show()
```

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16])
plt.show()
```

```python
#Parameter ketiga ini sepenuhnya mengadopsi gaya formatting pada MATLAB
#Nilai default untuk parameter  ketiga ini adalah 'b-'
#Pemanggilan fungsi axis() menyertakan sebuah list [xmin, xmax, ymin, ymax]
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
#Parameter ketiga pengatur bentuk marker yang dimaksud adalah karakter o bukan nol
```

```python
#Multiplot dalam satu pemanggilan fungsi plot()
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, 'r--') # 'r--' merepresentasikan warna merah
plt.plot(t, t**2, 'bs') # 'bs' merepresentasikan warna biru
plt.plot(t, t**3, 'g^') # 'g^' merepresentasikan warna hijau
plt.show()
```

```python
#Melakukan multi plot dalam satu pemanggilan fungsi plot()
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
        t, t**3, 'g^')
plt.show()
```

```python
#PLOTTING DENGAN KEYWORDS
#Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument(kwargs) seperti pada dictionary dan pandas DataFrame
```

```python
data = {'a' : np.arange(50),
        'c' : np.random.randint(0, 50, 50),
        'd' : np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a', 'b', data = data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c ='c', s='d', data = data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
