# Bar Plot / Bar Chart
Membuat bar plot dengan matplotlib

## Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Simple Bar Plot
* Kategori 'A' akan bersosialisasi dengan angka 25
* Kategori 'B' akan bersosialisasi dengan angka 45
* Kategori 'C' akan bersosialisasi dengan angka 55
* Kategari 'D' akan bersosialisasi dengan angka 125
* Kategori 'E' akan bersosialisais dengan angka 225

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data) # membuthkan sekitar 2 parameter
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### Pengaturan Grid dan Color

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori,
        data,
        color='red',  # Mengatur warna bar plot
        alpha = 0.75) # Mengatur transparasi pada bar plotnya

plt.grid(linestyle='--',
         linewidth=2, # Tebalnya garis
         axis ='x',   # Arah garis (Bisa x, y, atau dua"nya)
         alpha =0.50) # Mengatur transparansi dari garisnya

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### Horisontal Bar Plot
* `plt.barh` membuat horisontal Bar Plot
* Jika ingin urutan Bar Plot dari A ke E menggunakan `plt.barh(kategori[::-1], data[::-1])`
* Jika ingin urutan Bar Plot dari E Ke A menggunakan `plt.barh(kategori, data)`

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
#plt.barh(kategori, data)
plt.barh(kategori[::-1], data[::-1]) 

plt.grid(linestyle='--',
         linewidth=1, 
         axis ='x',   
         alpha =0.75) 

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
```

### Stacked Bar Plot
Disebut juga diagram batang bertumpuk
* **Data 1** dipresentasikan dalam warna biru
* **Data 2** dipresentasikan dalam warna jingga
* `bottom=data` untuk memilih data mana yang ditempatkan dibawah
* Jika`plt.bar(kategori, data1, label='Data1')` dan `plt.bar(kategori, data2, label='Data2')` dijalankan akan terjadi penimpaan ploting
* Jika `plt.bar(kategori, data1, label='Data1')` dan `plt.bar(kategori, data2, label='Data2', bottom=data1)` dijalankan tidak akan terjadi penimpaan ploting tetapi ditumpukan

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data1')
plt.bar(kategori, data2, label='Data2')
#plt.bar(kategori, data2, label='Data2', bottom=data1)

plt.legend()

plt.grid(linestyle='--',
         linewidth=1, 
         axis ='y',   
         alpha =0.70) 

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

### Grouped Bar Plot
* `plt.bar` untuk membuat bar plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))
x # Menampilkan array yang memiliki suatu deret bilangan dan deret itu ditentukan oleh nilai diatas
```

```python
width = 0.35 # Menentukan besar Bar Plotnya (semakin kecil nilai, semakin tipis barnya)

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)
plt.grid(linestyle='--',
         linewidth=1, 
         axis ='y',   
         alpha =0.75) 

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

# Pie Plot / Pie Chart
Membuat pie plot dengan matplotlib

## Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

### Simple Pie Plot
* `plt.pie` untuk membuat pie plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data,
        labels=kategori,    # Menambahkan label untuk memberi nama pada setiap area plot dengan list kategori diatas
        autopct='%1.1f%%',  # Berguna untuk menampilkan penanda presentase setiap areanya
        startangle=90)      # Berguna sebagai rotasi di pie plotnya

plt.title('Contoh Pie Plot')

plt.show()
```

### Pengaturan Legend dan Color
* `plt.legend()` berguna untuk menampilkan keterangan ploting

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        autopct='%1.1f%%',  
        startangle=90) 

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

### Exploded Pie Plot
* `explode=explode_var` ini akan menampilkan pie plot yang kesannya seperti ditarik keluar atau meledak"

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',  
        startangle=90) 

plt.legend()
plt.title('Contoh Exploded Pie Plot')

plt.show()
```

# Table Plot / Table Chart
Membuat table plot dengan matplotlib

## Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample Data

```python
data_uas=[
    ['Bejo', 70],
    ['Tejo', 83],
    ['Cecep', 62],
    ['Wati', 74],
    ['karti', 71]
]
```

```python
data_uas
```

### Table Plot dengan OO Style

```python
fig, ax = plt.subplots() #cara OO Style

table = plt.table(cellText=data_uas, loc='center')

table.set_fontsize(14)
table.scale(0.5, 4)     # Ukuran kolom, baris

ax.axis(False)

plt.show()
```

### Table Plot dengan pyplot Style

```python
table = plt.table(cellText=data_uas,
                  loc='center')

table.set_fontsize(14)
table.scale(1, 4)     # Ukuran kolom, baris

ax = plt.gca()
ax.axis(False)

plt.show()
```
