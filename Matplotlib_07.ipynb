## 3D Axes

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

## 3D Line Plot

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20* z)
y = z * np.cos(20* z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='red')

ax.set(title='Contoh 3D Line Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')
plt.show()
```

## 3D Scatter Plot

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20* z)
y = z * np.cos(20* z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='green')

ax.set(title='Contoh 3D Scatter Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Sample dataset

```python
n = np.linspace(-3.0, 3.0, 100)

x, y =np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## Contour Plot

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Filled Contour Plot
- Saat di run, Contour Plot akan berwarna

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Sample dataset

```python
n = np.linspace(-3.0, 3.0, 100)

x, y =np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## 3D Wireframe Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='lightgreen')

ax.set(title='3D WireFrame',
       xlabel='X',
       ylabel='Y',
       zlabel='Z')
plt.show()
```

## 3D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# ax.plot_surface(x, y, z, color='pink')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set(title='3D Surface Plot',
       xlabel='X',
       ylabel='Y',
       zlabel='Z')
plt.show()
```
